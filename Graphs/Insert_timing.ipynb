```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for MongoDB and PostgreSQL
# Replace these with your actual loading times
dataset_sizes = [25, 50, 75, 100]  # Representing percentages
#mongodb_load_times = [41.81, 82.34, 122.05, 160.89]  # Replace with MongoDB load times
#postgresql_load_times = [45.74, 84.60, 123.81, 163.66]  # Replace with PostgreSQL load times

mongodb_load_times = [43.82, 95.51, 150.14, 297.54]  # Replace with MongoDB load times
postgresql_load_times = [43.82, 109.69, 168.27, 227.59]  # Replace with PostgreSQL load times

# Convert percentages for plotting on x-axis
x_ticks = np.array(dataset_sizes)

plt.figure(figsize=(12, 7))

# Plot MongoDB load times
plt.plot(x_ticks, mongodb_load_times, marker='o', label='MongoDB', linestyle='-', linewidth=2, color='#1f77b4', markersize=8)

# Plot PostgreSQL load times
plt.plot(x_ticks, postgresql_load_times, marker='s', label='PostgreSQL', linestyle='--', linewidth=2, color='#ff7f0e', markersize=8)

# Customize the axes and grid
plt.xticks(x_ticks, [f"{x}%" for x in dataset_sizes])
plt.yticks(color='gray')
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Add title and labels with a font size
plt.title('SF Crime Data Load Time for MongoDB vs PostgreSQL', fontsize=16, fontweight='bold')
plt.xlabel('Dataset Size', fontsize=14)
plt.ylabel('Time in Seconds', fontsize=14)

# Add a legend with a shadow
plt.legend(loc='upper left', shadow=True)

# Add text annotations with the load time next to each point
for i, txt in enumerate(mongodb_load_times):
    plt.annotate(txt, (x_ticks[i], mongodb_load_times[i] + 10), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(postgresql_load_times):
    plt.annotate(txt, (x_ticks[i], postgresql_load_times[i] + 10), textcoords="offset points", xytext=(0,5), ha='center')

# Increase the bottom margin to prevent xlabel cutoff
plt.subplots_adjust(bottom=0.15)

# Display the plot
plt.show()
```
