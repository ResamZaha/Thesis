```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data for MongoDB, Cassandra, and MySQL
mongodb_times = [450, 200, 300, 400]  # Replace with actual data
cassandra_times = [550, 250, 350, 450]  # Replace with actual data
mysql_times = [600, 300, 400, 500]  # Replace with actual data

# Calculate the average for each database
mongodb_avg = np.mean(mongodb_times)
cassandra_avg = np.mean(cassandra_times)
mysql_avg = np.mean(mysql_times)

# Append the averages to the respective lists
mongodb_times.append(mongodb_avg)
cassandra_times.append(cassandra_avg)
mysql_times.append(mysql_avg)

# Define the figure size and style
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10, 6))

# The x location for the groups
ind = np.arange(len(mongodb_times))  # the x locations for the groups
width = 0.2  # the width of the bars

# Generate the bars for MongoDB, Cassandra, and MySQL
rects1 = ax.bar(ind - width, mongodb_times, width, label='MongoDB', color='#FFA07A')
rects2 = ax.bar(ind, cassandra_times, width, label='Cassandra', color='#20B2AA')
rects3 = ax.bar(ind + width, mysql_times, width, label='MySQL', color='#778899')

# Add labels and titles
ax.set_ylabel('Time in milliseconds', fontsize=12)
ax.set_xlabel('Queries', fontsize=12)
ax.set_title('Database Query Performance', fontsize=14)
ax.set_xticks(ind)
ax.set_xticklabels(('Query1', 'Query2', 'Query3', 'Query4', 'Average'), fontsize=10)
ax.legend(fontsize=10)

# Adding gridlines behind the bars
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')

# Function to attach a text label above each bar, displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(int(height)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

# Call the function to label the bars
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Show the plot with a tight layout
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_performance_graph(mongodb_times, sql_times, title, mongodb_label, sql_label, tick_labels):
    plt.style.use('ggplot')
    fig, ax = plt.subplots(figsize=(10, 6))

    ind = np.arange(len(mongodb_times))  # Adjusted for dynamic bar count
    width = 0.35

    rects1 = ax.bar(ind - width/2, mongodb_times, width, label=mongodb_label, color='#2ca02c')
    rects2 = ax.bar(ind + width/2, sql_times, width, label=sql_label, color='#1f77b4')

    ax.set_ylabel('Time in seconds')
    ax.set_xlabel('Queries')
    ax.set_title(title)
    ax.set_xticks(ind)
    ax.set_xticklabels(tick_labels, rotation=45)  # Adjust for better label readability
    ax.legend()

    ax.set_axisbelow(True)
    ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.3)

    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{:.2f}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    fig.tight_layout()
    plt.show()
```
