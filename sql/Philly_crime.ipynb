<h3> Detailed Analysis of Specific Crime Types</h3>

```python

start_time = time.time()

query = text("""
WITH CrimeDetails AS (
    SELECT 
        "text_general_code",
        "location_block",
        "hour"
    FROM philly_crime
    WHERE "text_general_code" IN ('Thefts', 'Fraud')
)
SELECT 
    "text_general_code",
    "location_block",
    COUNT(*) AS total_cases,
    AVG(hour) AS avg_time_of_crime
FROM CrimeDetails
GROUP BY "text_general_code", "location_block"
ORDER BY "text_general_code", total_cases DESC;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame as a table
display(df)

# Print execution time
print(f"Execution Time: {execution_time:.2f} seconds")
```
