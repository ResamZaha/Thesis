<h3> Monthly Average Temperature, Wind Speed with Condition Counts </h3>

```python

start_time = time.time()

# Corrected SQL query with the proper date format
query = text("""
WITH MonthlyData AS (
    SELECT 
        EXTRACT(MONTH FROM TO_DATE(valid_date_time, 'DD/MM/YYYY HH24:MI:SS')) AS month,
        AVG(temp) AS avg_temp,
        AVG(wspd) AS avg_wspd,
        COUNT(DISTINCT wx_phrase) AS distinct_conditions
    FROM sf_weather
    WHERE pressure BETWEEN 990 AND 1020
    GROUP BY month
)
SELECT month, avg_temp, avg_wspd, distinct_conditions
FROM MonthlyData
WHERE avg_wspd > 10
ORDER BY month;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Display the DataFrame as a table
display(df)  # If you are in a Jupyter notebook environment, otherwise use print(df)

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> Daily Data Analysis </h3>

```python

start_time = time.time()

# Corrected SQL query with the proper date format
query = text("""
SELECT 
    day,
    MAX(max_temp) - MIN(min_temp) AS temp_variation,
    MIN(vis) AS min_visibility
FROM (
    SELECT 
        TO_DATE(valid_date_time, 'DD/MM/YYYY HH24:MI:SS') AS day, 
        max_temp, 
        min_temp, 
        vis
    FROM sf_weather
) AS DailyData
GROUP BY day
HAVING MAX(max_temp) - MIN(min_temp) > 15 AND MIN(vis) < 5
ORDER BY temp_variation DESC;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Display the DataFrame as a table
display(df)  # If you are in a Jupyter notebook environment, otherwise use print(df)

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> Extreme weather days <h3>

```python
start_time = time.time()

# Corrected SQL query with the proper date format
query = text("""
SELECT 
    TO_DATE(valid_date_time, 'DD/MM/YYYY HH24:MI:SS') AS day, 
    MAX(temp) AS max_temp,
    MAX(wspd) AS max_wspd,
    MIN(vis) AS min_vis
FROM sf_weather
GROUP BY day
HAVING MAX(temp) > 30 OR MAX(wspd) > 50 OR MIN(vis) < 2
ORDER BY day;

""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Display the DataFrame as a table
display(df)  # If you are in a Jupyter notebook environment, otherwise use print(df)

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> Query 2.1.3 Identifying records where the wind chill is less than the actual temperature and the wind speed is above a certain level</h3>

```python
start_time = time.time()

# Modified SQL query
query = text("""
SELECT 
    "Unnamed: 0",
    "valid_time_gmt",
    "temp",
    "wc",
    "wspd",
    "temp" - "wc" AS wind_chill_gap,
    CASE
        WHEN "wc" < "temp" THEN 'Wind Chill Lower'
        ELSE 'Temperature Lower or Equal'
    END AS temperature_comparison,
    CASE
        WHEN "wspd" > 25 THEN 'High Wind Speed'
        ELSE 'Normal Wind Speed'
    END AS wind_speed_status,
    "wx_phrase",
    "valid_date_time"
FROM 
    sf_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.137 seconds
```

<h3> Query 2.1.4 Checking for temperature variations and categorizing weather conditions based on wind speed and temperature </h3>

```python
start_time = time.time()

# Modified SQL query
query = text("""
SELECT 
    "Unnamed: 0",
    "valid_time_gmt",
    "max_temp" - "min_temp" AS temp_range,
    "wspd" + 10 AS adjusted_wspd,
    CASE
        WHEN "wspd" > 20 THEN 'Windy'
        WHEN "temp" < 32 THEN 'Freezing'
        ELSE 'Mild'
    END AS weather_condition,
    "wx_phrase",
    "valid_date_time"
FROM 
    sf_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.131 seconds
```

<h3> Query 2.1.5 Comparing day and night temperatures and visibility for weather predictions </h3>

```python
start_time = time.time()

# Modified SQL query
query = text("""
SELECT 
    "Unnamed: 0",
    "valid_time_gmt",
    "max_temp" + "min_temp" AS combined_temperature,
    "vis" / 2 AS half_visibility,
    CASE
        WHEN "max_temp" > 75 AND "min_temp" < 50 THEN 'Wide Temperature Range'
        WHEN "vis" < 5 THEN 'Foggy'
        ELSE 'Clear'
    END AS weather_prediction,
    "wx_phrase",
    "valid_date_time"
FROM 
    sf_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.125 seconds
```
