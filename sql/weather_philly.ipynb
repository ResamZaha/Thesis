<h3> total sum of temperatures and pressures for each distinct weather phrase (wx_phrase) </h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    wx_phrase,
    SUM(temp) AS total_temp,
    SUM(pressure) AS total_pressure
FROM philly_weather
GROUP BY wx_phrase
ORDER BY wx_phrase;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> Daily Data</h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    TO_DATE(valid_date_time, 'DD/MM/YYYY') AS day,
    AVG(temp) AS avg_temp,
    SUM(wspd) AS total_wspd,
    MAX(vis) AS max_visibility,
    COUNT(DISTINCT wx_phrase) AS distinct_weather_conditions
FROM philly_weather
GROUP BY day
ORDER BY day;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> 6.1.1 Selecting records with temperature differences within a specified range and wind speed above a certain threshold </h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    "Unnamed: 0",
    "expire_time_gmt",
    "valid_time_gmt",
    "temp",
    "wspd",
    "wx_phrase",
    "valid_date_time",
    "expire_date_time",
    "max_temp" - "min_temp" AS temperature_difference,
    CASE 
        WHEN "wspd" > 20 THEN 'High wind'
        ELSE 'Low wind'
    END AS wind_speed_category
FROM 
    philly_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.144 seconds
```

<h3> 6.1.2 Identifying records where visibility is within a certain range and the temperature is below freezing </h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    "Unnamed: 0",
    "expire_date_time",
    "expire_time_gmt",
    "valid_date_time",
    "valid_time_gmt",
    "vis",
    "wx_phrase",
        CASE 
        WHEN "wdir" BETWEEN 0 AND 180 THEN 'Easterly Wind'
        ELSE 'Westerly Wind'
    END AS wind_direction,
    "temp",
    CASE 
        WHEN "temp" < 0 THEN 'Below freezing'
        ELSE 'Above freezing'
    END AS temperature_status
FROM 
    philly_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.137 seconds
```

<h3> 6.1.3 Comparing temperature differences with visibility ranges to classify weather severity</h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    "Unnamed: 0",
    "valid_time_gmt",
    "temp" - "wc" AS realfeel_difference,
    "vis",
    CASE
        WHEN "vis" < 5 THEN 'Low Visibility'
        WHEN "vis" BETWEEN 5 AND 15 THEN 'Moderate Visibility'
        ELSE 'High Visibility'
    END AS visibility_status,
    CASE
        WHEN "temp" - "wc" > 20 THEN 'Significant Wind Chill Effect'
        ELSE 'Negligible Wind Chill Effect'
    END AS wind_chill_effect,
    "wx_phrase",
    "valid_date_time"
FROM 
    philly_weather;
""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#0.139 seconds
```

<h3> 6.1.4 Calculating wind direction variance and assigning weather status based on wind chill </h3>

```python
start_time = time.time()

# SQL query with SUM
query = text("""
SELECT 
    "Unnamed: 0",
    "valid_time_gmt",
    "wdir" + 180 AS opposite_wdir,
    "wc" - 5 AS reduced_wind_chill,
    "temp" * "wspd" AS wind_temperature_factor,
    ("temp" + "wspd") / 2 AS comfort_index,
    CASE
        WHEN "temp" > 85 THEN 'Too Hot'
        WHEN "temp" < 32 THEN 'Too Cold'
        ELSE 'Comfortable'
    END AS comfort_status,
    CASE
        WHEN "wc" < 0 THEN 'Extreme Cold'
        WHEN "wc" BETWEEN 0 AND 32 THEN 'Chilly'
        ELSE 'Not Chilly'
    END AS wind_chill_status,
    "wx_phrase",
    "valid_date_time"
FROM 
    philly_weather;

""")

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(query, engine)

end_time = time.time()
execution_time = end_time - start_time

# Print the DataFrame
display(df)

print(f"Execution Time: {execution_time:.2f} seconds")
#.142 seconds
#.331 seconds
"""{
			"_id": 0,      "Unnamed: 0": 1,
      "valid_time_gmt": 1,
      "opposite_wdir": { "$add": ["$wdir", 180] },
      "reduced_wind_chill": { "$subtract": ["$wc", 5] },
      "wind_temperature_factor": { "$multiply": ["$temp", "$wspd"] },
      "comfort_index": { "$divide": [{ "$add": ["$temp", "$wspd"] }, 2] },
      "comfort_status": {
        "$switch": {
          "branches": [
            { "case": { "$gt": ["$temp", 85] }, "then": "Too Hot" },
            { "case": { "$lt": ["$temp", 32] }, "then": "Too Cold" }
          ],
          "default": "Comfortable"
        }
      },
      "wind_chill_status": {
        "$switch": {
          "branches": [
            { "case": { "$lt": ["$wc", 0] }, "then": "Extreme Cold" },
            { "case": { "$and": [{ "$gte": ["$wc", 0] }, { "$lte": ["$wc", 32] }] }, "then": "Chilly" }
          ],
          "default": "Not Chilly"
        }
      },
      "wx_phrase": 1,
      "valid_date_time": 1
    }"""
```
