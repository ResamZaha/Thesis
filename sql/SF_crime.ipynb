<h2><b> Per Year Crime using Pipeline</h2></b>

```python
start_time = time.time()

query = text("""
SELECT EXTRACT(YEAR FROM TO_DATE("Date", 'MM/DD/YYYY')) AS "Year", COUNT(*) AS crime_count
FROM sf_crimedata
GROUP BY EXTRACT(YEAR FROM TO_DATE("Date", 'MM/DD/YYYY'))
ORDER BY "Year";
""")

with engine.connect() as con:
    rs = con.execute(query)
    for row in rs:
        # Depending on the database and driver, you might need to access the row as a dictionary
        print(f"Year: {int(row['Year'])}, Crime Count: {row['crime_count']}")

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h2><b> Per Month Crime using Pipeline</h2></b>

```python
start_time = time.time()

query = text("""
SELECT 
    EXTRACT(MONTH FROM TO_DATE("Date", 'MM/DD/YYYY')) AS month,
    COUNT(*) AS crime_count
FROM sf_crimedata
GROUP BY month
ORDER BY month;
""")

with engine.connect() as con:
    rs = con.execute(query)
    for row in rs:
        # Formatting month for display
        month_name = time.strftime('%B', time.strptime(str(int(row['month'])), "%m"))
        print(f"Month: {month_name}, Crime Count: {row['crime_count']}")

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution Time: {execution_time:.2f} seconds")
```

<h3> Year-over-Year Change in Crime Types within Vulnerable Areas </h3>

```python
import pandas as pd
from sqlalchemy import create_engine, text
import time

start_time = time.time()

query = text("""
WITH YearlyCrimeStats AS (
    SELECT 
        "Areas of Vulnerability, 2016 2 2" AS "Vulnerability_Level",
        "Category",
        EXTRACT(YEAR FROM TO_DATE("Date", 'MM/DD/YYYY')) AS "Year",
        COUNT(*) AS "Total_Incidents"
    FROM sf_crimedata
    WHERE "Areas of Vulnerability, 2016 2 2" IS NOT NULL
    GROUP BY 1, 2, 3
)
SELECT 
    cs."Vulnerability_Level",
    cs."Category",
    cs."Year" AS "Current_Year",
    py."Total_Incidents" AS "Previous_Year_Incidents",
    cs."Total_Incidents" AS "Current_Year_Incidents",
    (cs."Total_Incidents" - COALESCE(py."Total_Incidents", 0)) AS "Year_Over_Year_Change"
FROM YearlyCrimeStats cs
LEFT JOIN YearlyCrimeStats py ON cs."Vulnerability_Level" = py."Vulnerability_Level"
    AND cs."Category" = py."Category"
    AND cs."Year" = py."Year" + 1
ORDER BY cs."Vulnerability_Level", cs."Category", cs."Year";

""")

# Assuming 'engine' is already defined and connected to your database
df = pd.read_sql_query(query, engine)

end_time = time.time()

# Display the DataFrame
display(df)  # If using Jupyter notebooks, otherwise use print(df)

print(f"Execution Time: {end_time - start_time:.2f} seconds")
```

<h3> Crime Incident Density by District and Time of Day <h3>

```python
start_time = time.time()

query = text("""
SELECT 
    "PdDistrict",
    CASE 
        WHEN EXTRACT(HOUR FROM TO_TIMESTAMP("Time", 'HH24:MI')) BETWEEN 0 AND 7 THEN 'Night'
        WHEN EXTRACT(HOUR FROM TO_TIMESTAMP("Time", 'HH24:MI')) BETWEEN 8 AND 15 THEN 'Day'
        ELSE 'Evening'
    END AS TimeOfDay,
    COUNT(*) AS Total_Incidents,
    AVG("X") AS Avg_Longitude,
    AVG("Y") AS Avg_Latitude
FROM sf_crimedata
GROUP BY "PdDistrict", TimeOfDay
ORDER BY "PdDistrict", TimeOfDay;
""")

df = pd.read_sql_query(query, engine)

end_time = time.time()

display(df)  # Use print(df) if not in a Jupyter notebook

print(f"Execution Time: {end_time - start_time:.2f} seconds")
```

<h3> Analysis of Crime Resolutions in Relation to Special Zones and Neighborhoods <h3>

```python
start_time = time.time()

query = text("""
SELECT 
    "Category",
    "Resolution",
    "Analysis Neighborhoods 2 2" AS Neighborhood,
    "Central Market/Tenderloin Boundary 2 2" AS Tenderloin_Boundary,
    "HSOC Zones as of 2018-06-05 2 2" AS HSOC_Zone,
    COUNT(*) AS Total_Cases,
    AVG("X") AS Avg_Longitude,
    AVG("Y") AS Avg_Latitude
FROM sf_crimedata
WHERE "Resolution" IN ('ARREST, BOOKED', 'NONE')
AND ("Central Market/Tenderloin Boundary 2 2" IS NOT NULL OR "HSOC Zones as of 2018-06-05 2 2" IS NOT NULL)
GROUP BY "Category", "Resolution", Neighborhood, Tenderloin_Boundary, HSOC_Zone
ORDER BY Total_Cases DESC;
""")

df = pd.read_sql_query(query, engine)

end_time = time.time()

display(df)  

print(f"Execution Time: {end_time - start_time:.2f} seconds")
```
