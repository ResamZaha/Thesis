<h3> Crime Incident Density by District and Time of Day <h3>

````python
# Record the start time
start_time = time.time()

# Aggregation pipeline to calculate the average number of cases per year
pipeline = [
    {
        "$addFields": {
            "Hour": {
                "$toInt": {
                    "$substr": ["$Time", 0, 2]  # Extracts the hour part from the "Time" field
                }
            }
        }
    },
    {
        "$addFields": {
            "TimeOfDay": {
                "$switch": {
                    "branches": [
                        {"case": {"$lt": ["$Hour", 8]}, "then": "Night"},
                        {"case": {"$and": [{"$gte": ["$Hour", 8]}, {"$lt": ["$Hour", 16]}]}, "then": "Day"}
                    ],
                    "default": "Evening"
                }
            }
        }
    },
    {
        # Stage to group by PdDistrict and TimeOfDay
        "$group": {
            "_id": {
                "PdDistrict": "$PdDistrict",
                "TimeOfDay": "$TimeOfDay"
            },
            "Total_Incidents": {"$sum": 1},
            "Avg_Longitude": {"$avg": "$X"},
            "Avg_Latitude": {"$avg": "$Y"}
        }
    },
    {
        # Stage to sort the results
        "$sort": {
            "_id.PdDistrict": 1,
            "_id.TimeOfDay": 1
        }
    },
    {
        # Stage to project the final structure
        "$project": {
            "_id": 0,
            "PdDistrict": "$_id.PdDistrict",
            "TimeOfDay": "$_id.TimeOfDay",
            "Total_Incidents": 1,
            "Avg_Longitude": 1,
            "Avg_Latitude": 1
        }
    }
]


# Execute the pipeline
# results = db.collection.aggregate(pipeline)

# Note: MongoDB does not directly support self-joins or easily calculating differences between documents in a single aggregation pipeline.
# For the year-over-year comparison, consider retrieving the data first and then performing the comparison in your application logic,
# or restructuring your data to facilitate easier comparisons within the database.

# Run the aggregation query
data = db.crime.sf.aggregate(pipeline)

# Collect the results into a list of dictionaries
results = list(data)

# Record the end time
end_time = time.time()
execution_time = end_time - start_time

# Create a DataFrame from the results
df = pd.DataFrame(results)

# Print the DataFrame as a table
display(df)  # Use print(df) if not in a Jupyter notebook

# Print execution time
print(f"Execution Time: {execution_time:.2f} seconds")
#1.4 seconds
````
