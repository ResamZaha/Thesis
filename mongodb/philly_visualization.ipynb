````python
start_time = time.time()

# MongoDB aggregation pipeline
pipeline = [
    {
        "$sort": {
            "location_block": -1 # Sort by location_block descending before grouping to get the "max"
        }
    },
    {
        "$group": {
            "_id": {
                "hour": "$hour",
                "text_general_code": "$text_general_code"
            },
            "total_incidents": { "$sum": 1 },
            "common_location": { "$first": "$location_block" } # After sorting, first will be the "max"
        }
    },
    {
        "$addFields": {
            "hour": "$_id.hour",
            "text_general_code": "$_id.text_general_code"
        }
    },
    {
        "$sort": {
            "hour": 1,
            "total_incidents": -1
        }
    },
    {
        "$project": {
            "_id": 0,
            "hour": 1,
            "text_general_code": 1,
            "total_incidents": 1,
            "common_location": 1
        }
    }
]

# Execute the aggregation pipeline
data = db.crime.philadelphia.aggregate(pipeline)

# Collect the results into a list of dictionaries
results = list(data)

# Record the end time
end_time = time.time()
execution_time = end_time - start_time

# Create a DataFrame from the results
df = pd.DataFrame(results)

# Print the DataFrame as a table
display(df)  # Use print(df) if not in a Jupyter notebook

# Print execution time
print(f"Execution Time: {execution_time:.2f} seconds")
#6.9 seconds
````
