<h3> 3.3.1 Annual Crime Summary by Type </h3>

````python
start_time = time.time()

# MongoDB aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": {
                "Year": "$Year",
                "PrimaryType": "$Primary Type"
            },
            "total_incidents": {"$sum": 1},
            "total_arrests": {
                "$sum": {
                    "$cond": [{"$eq": ["$Arrest", True]}, 1, 0]
                }
            },
            # Attempt to capture the most common location description
            "common_locations": {
                "$push": "$Location Description"
            }
        }
    },
    {
        "$addFields": {
            "common_location": {
                "$arrayElemAt": ["$common_locations", 0]  # Simplistic approach
            }
        }
    },
    {
        "$project": {
            "total_incidents": 1,
            "total_arrests": 1,
            "common_location": 1
        }
    },
    {
        "$sort": {"_id.Year": 1, "total_incidents": -1}
    }
]

# Execute the aggregation pipeline
data = db.crime.chicago.aggregate(pipeline)

# Collect the results into a list of dictionaries
results = []
for doc in data:
    result = {
        "Year": doc['_id']['Year'],
        "Primary Type": doc['_id']['PrimaryType'],
        "Total Incidents": doc['total_incidents'],
        "Total Arrests": doc['total_arrests'],
        "Common Location": doc.get('common_location')  # Adjust based on your logic to determine the common location
    }
    results.append(result)

# Create a DataFrame from the results
df = pd.DataFrame(results)

# Record the end time
end_time = time.time()
execution_time = end_time - start_time

# Display the DataFrame as a table
display(df)  


print(f"Execution Time: {execution_time:.2f} seconds")
#6.9 seconds
````

````python

start_time = time.time()

# MongoDB aggregation pipeline
pipeline = [
    # Stage 1: Group by Description and Year
    {
        "$group": {
            "_id": {
                "Description": "$Description",
                "Year": "$Year"
            },
            "count": {"$sum": 1},
            "arrest_count": {
                "$sum": {
                    "$cond": [{"$eq": ["$Arrest", True]}, 1, 0]
                }
            }
        }
    },
    # Stage 2: Filter groups with more than 50 records
    {
        "$match": {
            "count": {"$gt": 50}
        }
    },
    # Stage 3: Project to format the output
    {
        "$project": {
            "_id": 0,
            "Description": "$_id.Description",
            "Year": "$_id.Year",
            "count": 1,
            "arrest_count": 1
        }
    },
    # Stage 4: Sort the results by Year and count in descending order
    {
        "$sort": {
            "Year": 1,
            "count": -1
        }
    }
]
# Execute the aggregation pipeline
data = db.crime.chicago.aggregate(pipeline)

# Collect the results into a list of dictionaries
results = list(data)

# Record the end time
end_time = time.time()
execution_time = end_time - start_time

# Create a DataFrame from the results
df = pd.DataFrame(results)

# Print the DataFrame as a table
display(df)  # Use print(df) if not in a Jupyter notebook

# Print execution time
print(f"Execution Time: {execution_time:.2f} seconds")
#5.6 seconds
````
