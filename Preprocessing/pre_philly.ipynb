```python
ROOTPATH = '/Users/nuhash/Desktop/Code/Thesis/'
PATH_DATA_FOLDER = ROOTPATH + 'data_philly/'

# list all the files in the folder
import os
files = os.listdir(PATH_DATA_FOLDER)
print(files)
```

```python
dfs = []
for file in files:
    if(file.endswith('(CSV).csv') == False):
        continue
    print(file)
    df = pd.read_csv(PATH_DATA_FOLDER + file)
    dfs.append(df)

# Concatenate all data into one DataFrame
df = pd.concat(dfs, ignore_index=True)
df.tail()
    
```

```python
# 2014-09-15 04:00:00+00
df['dispatch_date_time'] = pd.to_datetime(
    df['dispatch_date_time'], format='%Y-%m-%d %H:%M:%S+00')
df = df.sort_values(by=['dispatch_date_time'])
```

```python

df.to_csv(ROOTPATH + 'philly_2006_2023.csv', index=False)
```
